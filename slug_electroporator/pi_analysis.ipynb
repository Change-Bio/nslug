```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Build dataframe from plate reader data (535-16/617-20)
records = []

# Controls
for val in [4622, 4620]:
    records.append({"condition": "PI", "group": "Control", "fluorescence": val})
for val in [491, 471, 484]:
    records.append({"condition": "Alive-PI", "group": "Alive", "fluorescence": val})
for val in [12490, 13571, 12742]:
    records.append({"condition": "Alive+PI", "group": "Alive", "fluorescence": val})
for val in [620, 592, 607]:
    records.append({"condition": "Dead-PI", "group": "Dead", "fluorescence": val})
for val in [29936, 26370, 28849]:
    records.append({"condition": "Dead+PI", "group": "Dead", "fluorescence": val})

# Voltage series: 1.0-2.0 kV in 0.2 steps, then 2.5 kV
voltages = [1.0, 1.2, 1.4, 1.6, 1.8, 2.0, 2.5]
row_b = [14651, 13943, 14274, 14611, 14794, 14645, 11664]
row_c = [11998, 14863, 15664, 13405, 13784, 15649, 11509]

for v, b, c in zip(voltages, row_b, row_c):
    records.append({"condition": f"{v} kV", "group": "Voltage", "fluorescence": b, "voltage_kV": v, "rep": "B"})
    records.append({"condition": f"{v} kV", "group": "Voltage", "fluorescence": c, "voltage_kV": v, "rep": "C"})

df = pd.DataFrame(records)
df
```

```python
# Controls plot
df_ctrl = df[df["group"] != "Voltage"]

fig, ax = plt.subplots()
sns.stripplot(data=df_ctrl, x="condition", y="fluorescence", hue="group",
              dodge=False, jitter=0.15, size=8, alpha=0.7, ax=ax)

# Mean lines
for i, cond in enumerate(df_ctrl["condition"].unique()):
    m = df_ctrl[df_ctrl["condition"] == cond]["fluorescence"].mean()
    ax.hlines(m, i - 0.3, i + 0.3, color="black", linewidth=2, zorder=5)

ax.set_ylabel("Fluorescence (AU)", fontweight='bold')
ax.set_xlabel("")
ax.tick_params(axis='x', rotation=45)
for label in ax.get_xticklabels():
    label.set_fontweight('bold')
plt.tight_layout()
fig.savefig("controls_plot.png", dpi=300)
plt.show()
```

```python
# Voltage series plot
df_volt = df[df["group"] == "Voltage"]

fig, ax = plt.subplots()
sns.stripplot(data=df_volt, x="condition", y="fluorescence",
              jitter=0.15, size=8, alpha=0.7, color="#377eb8", ax=ax)

# Mean lines
for i, cond in enumerate(df_volt["condition"].unique()):
    m = df_volt[df_volt["condition"] == cond]["fluorescence"].mean()
    ax.hlines(m, i - 0.3, i + 0.3, color="black", linewidth=2, zorder=5)

ax.set_ylabel("Fluorescence (AU)", fontweight='bold')
ax.set_xlabel("Voltage", fontweight='bold')
ax.tick_params(axis='x', rotation=45)
for label in ax.get_xticklabels():
    label.set_fontweight('bold')
plt.tight_layout()
fig.savefig("voltage_series_plot.png", dpi=300)
plt.show()
```

```python
# Combined plot - all conditions on one axis
order = ["PI", "Alive-PI", "Alive+PI", "Dead-PI", "Dead+PI"] + [f"{v} kV" for v in voltages]
palette = {"Control": "#999999", "Alive": "#4daf4a", "Dead": "#e41a1c", "Voltage": "#377eb8"}

fig, ax = plt.subplots()
sns.stripplot(data=df, x="condition", y="fluorescence", hue="group",
              order=order, palette=palette,
              dodge=False, jitter=0.15, size=8, alpha=0.7, ax=ax)

# Mean lines
for i, cond in enumerate(order):
    m = df[df["condition"] == cond]["fluorescence"].mean()
    ax.hlines(m, i - 0.3, i + 0.3, color="black", linewidth=2, zorder=5)


ax.set_yticks([0,15000,30000])
ax.set_yticklabels([0,15,30])

ax.set_ylabel("Fluorescence (AUx10$^3$)", fontweight='bold')
ax.set_xlabel("")
ax.legend().remove()
ax.tick_params(axis='x', rotation=45)
for label in ax.get_xticklabels():
    label.set_fontweight('bold')
plt.tight_layout()
fig.savefig("combined_plot.png", dpi=300)
plt.show()
```
